```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Signal parameters
amplitude = 1
wc = 2*np.pi*4  
wm = 2*np.pi*1
delta_w = 4

vitesse_anim = 0.1

# Plot settings
fig, ax = plt.subplots()
ax.axis('equal') 
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

# Initialisation du point sur le cercle et de la ligne de projection
point, = ax.plot([], [], 'bo')  # point sur le cercle
ligne, = ax.plot([], [], 'r-')  # ligne projetée pour l'amplitude du sinus
points_x, points_y = [], []  # Listes pour stocker les coordonnées des points


# Initialisation de l'animation
def init():
    point.set_data([], [])
    ligne.set_data([], [])
    return point, ligne

# Fonction d'animation qui sera appelée à chaque frame
def animate(i):
    phase_c = wc*i*vitesse_anim
    phase_m = wm*i*vitesse_anim
    phasor = amplitude * np.exp(1j * (phase_c)) * np.exp(1j * delta_w *(np.cos(phase_m)))
    
    x = np.real(phasor)
    y = np.imag(phasor)
    points_x.append(x)
    points_y.append(y)

     # Mise à jour des données du point et de la ligne
    point.set_data(points_x, points_y)
    
     # Mise à jour des données de la ligne avec la liste des points
    ligne.set_data([x,0], [y,0])
    
    return point, ligne

# Création de l'animation
ani = FuncAnimation(fig, animate, init_func=init, frames=np.arange(0, 10, 0.1), interval=20, blit=True)
```

```python
ani.save('animation2.gif', writer='Pillow', fps=20)
```
